# Task-3-Socialgraph MapReduce

```python
%%file python/socialgraph-3.py
#!/usr/bin/env python3

from mrjob.job import MRJob
from mrjob.step import MRStep
import heapq


class Count(MRJob):
    
    def mapper(self, _, value):

        X=value.split("\t")
        yield(X[1], 1)

    def combiner(self, key, value):
        yield(key, sum(value))
 

    def reducer_init(self):
        self.heap = []
        
#Heap queue algorithm
#ref: https://docs.python.org/3/library/heapq.html
        
    def reducer(self, key, value):
        heapq.heappush(self.heap,(sum(value), key))
        if len(self.heap) > 100:
            heapq.heappop(self.heap)
            
    def reducer_final(self):
        for (value, key) in self.heap:
            yield(key, value)
            
    def top_mapper(self, key, value):
        yield (str(100), (value, key))
        
    def top_reducer(self,_,value):
        for value in heapq.nlargest(100, value):
            yield(value[1], value[0])

            
    def steps(self):
        return [
            MRStep(mapper=self.mapper,
                  combiner=self.combiner,
                  reducer_init=self.reducer_init,
                  reducer=self.reducer,
                  reducer_final=self.reducer_final
                  ),
            
            MRStep(mapper=self.top_mapper,
                  reducer=self.top_reducer) ]
    
  
if __name__ == '__main__':
    Count.run()
```
